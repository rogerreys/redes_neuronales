```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sigmoid
def sigmoid(x):
    return 1/(1+np.exp(-x))

x= np.linspace(-10, 10, 100)
y = sigmoid(x)

plt.plot(x,y)
```

```python
# Escalonado
def step(x):
    return np.piecewise(x, [x<0.0, x>0.0], [0,1])
    # return np.array([1 if i>0 else 0 for i in x])

x = np.linspace(-10, 10, 100)
y = step(x)
plt.plot(x,y)
```

```python
# relu(x)
def relu(x):
    return np.piecewise(x, [x<0.0, x>0.0], [0,lambda x:x])
    return np.array([i if i>0 else 0 for i in x])

x = np.linspace(-10, 10, 100)
y = relu(x)
plt.plot(x,y)
```

```python
# tanh(x)
def tanh(x):
    return np.tanh(x)

x = np.linspace(-10, 10, 100)
y = tanh(x)
plt.plot(x,y)
```
