# Manejo de imagenes

1. Para procesar imagenes se debe transformar en grises siempre y cuando el color no importe
2. Maneja una escala de tama√±o predeterminado, manejar las mismas dimensiones
3. Usar Data augmentation, se puede crear mas set de datos con la misma imagen

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
```

```python
im = io.imread("./src/cnn-data-sources/Mara.jpeg")
im.shape # (800, 600, 3) tiene tres capas de color
```

```python
plt.imshow(im)
```

```python
r=im[:,:,0]
g=im[:,:,1]
b=im[:,:,2]

# plt.imshow(g)
plt.imshow(g.T,cmap="gray")
```

```python
print(r.shape)
aux_dim = np.zeros([800,600])
red = np.dstack((r, aux_dim, aux_dim)).astype(np.uint16)
green = np.dstack((aux_dim, g, aux_dim)).astype(np.uint16)
blue = np.dstack(( aux_dim, aux_dim, b)).astype(np.uint16)
print(red.shape, green.shape, blue.shape)
plt.imshow(green)
plt.imshow(blue)
```
